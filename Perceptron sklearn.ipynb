```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale
```

```python
#coverting labels to numbers
def labels_conversion(s):
    s=str(s).lower()
    if s == 'm':
        return 1
    elif s=='b':
        return 0
    else: 
        return 'X'
```

```python
dataset=pd.read_csv('/Users/kaku/Downloads/AIML/git/Datasets/Wisconsin/wdbc.data.txt',header=None, converters={1:labels_conversion})
```

```python
dataset.head()
#type(dataset[30][1])
```

```python
dataset=dataset.drop([0],axis=1)
```

```python
dataset.head()
```

```python
columns=dataset.columns.tolist()
columns=columns[1:]+columns[0:1]
columns
```

```python
dataset=dataset[columns]
dataset.head()
```

```python
dataset=dataset.T.reset_index(drop=True).T
dataset.head()
```

```python
'X' in dataset[30]
```

```python
#dataset.describe()
dataset.count()
```

```python
#data spilt
data=dataset.values[:,:30]
labels=dataset.values[:,-1]
x_train, x_test, y_train, y_test= train_test_split(data,labels,test_size=0.35,random_state=100)
#len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
#standardisation
x_train=scale(x_train)
x_test=scale(x_test)
```

```python
clf=linear_model.Perceptron(max_iter=100,eta0=0.001)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
